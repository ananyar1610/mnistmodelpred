```python
import pandas as pd
```

```python
train=pd.read_csv('mnist_train.csv')
test=pd.read_csv('mnist_test.csv')
```

```python
train.columns
```

```python
train.head()
```

```python
train.shape
```

```python
train.drop_duplicates(inplace=True)
test.drop_duplicates(inplace=True)
```

```python
train.dropna(inplace=True)
test.dropna(inplace=True)
```

```python
y_train=train['label']
```

```python
x_train=train.drop(columns='label')
```

```python
y_test=test['label']
```

```python
x_test=test.drop(columns='label')
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train.head()
```

```python
x_train
```

```python
y_train.head()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train, y_train)
pred=model.predict(x_test)
```

```python
pred[:10]
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test, pred)
r2
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train, y_train)
pred=model.predict(x_test)
```

```python
r2_log=r2_score(y_test, pred)
r2_log
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, pred)
acc
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(x_train, y_train)
knn_pred=model.predict(x_test)
```

```python
#accuracy score
import numpy as np
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, knn_pred)
acc
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score
print(f"Precision: {precision_score(y_test, knn_pred, average='weighted')}")
print(f"Recall: {recall_score(y_test, knn_pred, average='weighted')}")
```

```python
#We also need to do Decision Tree, Random Forest and XGBoost
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train, y_train)
dct_pred=model.predict(x_test)

```

```python
acc2=accuracy_score(y_test,dct_pred)
print(f"Acuuracy of Decision Tree is {acc2}")
```

```python
#Random Forest and it relevant score calculations
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train, y_train)
rfc_pred=model.predict(x_test)
```

```python
print(rfc_pred)
```

```python
#Random Forest scores
acc3=accuracy_score(y_test,rfc_pred)
print(f"Accuracy of Random Forest is {acc3}")
rfc_precision=precision_score(y_test,rfc_pred, average='weighted')
print(f"Precision of Random Forest is {rfc_precision}")
rfc_recall=recall_score(y_test,rfc_pred, average='weighted')
print(f"Recall of Random Forest is {rfc_recall}")
```

```python
#XG BOOST
from xgboost import XGBClassifier
model=XGBClassifier()
model.fit(x_train, y_train)
xgb_pred=model.predict(x_test)
```

```python
#XGB Scores
print(f"Accuracy of XCG Boost is {accuracy_score(y_test, xgb_pred)}")
print(f"Precision of XGBoost is {precision_score(y_test, xgb_pred, average='weighted')}")
print(f"Recall of XGBoost is {recall_score(y_test, xgb_pred, average='weighted')}")
```

XG Boost Seems to be the most efficient model in terms of Accuracy, Precision and Recall.
